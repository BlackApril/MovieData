# **데이터 불러오기**

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import linear_kernel
url1 = ('https://raw.githubusercontent.com/sd12832/MoViZ/master/' +
        'src/tmdb_5000_credits.csv')
df1 = pd.read_csv(url1)
print(df1.shape)
df1.head()
```

```python
url2 = ('https://raw.githubusercontent.com/sd12832/MoViZ/master/' +
        'src/tmdb_5000_movies.csv')
df2 = pd.read_csv(url2)
print(df2.shape)
df2.head(5)
```

# **데이터 전처리**

```python
# prompt: df1의 컬럼명을['id', 'tittle', 'cast', 'crew']로 변경

df1.columns = ['id', 'tittle', 'cast', 'crew']
```

```python
# prompt: df2와 df1을 'id' 컬럼을 기준으로 병합 (merge)

df1_df2_merge = df1.merge(df2, on='id')
print(df1_df2_merge.shape)
df1_df2_merge.head()
```

# **영화 추천 시스템을 위한 IMDB 가중평점 계산 코드**

```python
# 평점 통계율 계산
C = df1_df2_merge['vote_average'].mean()
m = df1_df2_merge['vote_count'].quantile(0.9)
C, m
```

```python
# 평점 투표수가 최소 투표수 보다 큰 경우만 q_movies로 저장
q_movies = df1_df2_merge.copy().loc[df1_df2_merge['vote_count'] >= m]
q_movies.shape
```

```python
# prompt: q_movies 데이터 프레임에 IDMB 평점 계산 함수 적용
#IDMB 평점 계산 함수
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    return (v/(v+m) * R) + (m/(m+v) * C)

# q_movies 데이터프레임에 IDMB 평점 계산 함수 적용
q_movies['score'] = q_movies.apply(weighted_rating, axis=1)
q_movies = q_movies.sort_values('score', ascending=False)
q_movies[['title', 'vote_count', 'vote_average', 'score','original_language']].head(10)
```

```python
#병합된 테이블에서 original_language가 'en'이면 f_movies로 저장
f_movies = df1_df2_merge.loc[df1_df2_merge['original_language'] != 'en']
f_movies.shape
```

```python
#평점 통계율 계산
C = f_movies['vote_average'].mean()
m = f_movies['vote_count'].quantile(0.5)
C,m
```

```python
# prompt: f_movies 데이터 프레임의 각 행에 weighted_rating 함수 생성

f_movies['score'] = f_movies.apply(weighted_rating, axis=1)

f_movies = f_movies.sort_values('score', ascending=False)

f_movies[['title', 'vote_count', 'vote_average', 'score','original_language']].head(10)
```

# **영어 불용어 제거**

```python
#영어 불용어 제거
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words ='english')
```

# **영어권 영화 Top 10**

```python
# prompt: 'original_language'가 'en'인 영화 Top 10을 추천해줘

# Filter movies with original_language as 'en'
en_movies = df1_df2_merge.copy().loc[df1_df2_merge['original_language'] == 'en']

# Calculate C and m for English movies
C_en = en_movies['vote_average'].mean()
m_en = en_movies['vote_count'].quantile(0.9)

# Apply the weighted_rating function to English movies
en_movies['score'] = en_movies.apply(lambda x: weighted_rating(x, m=m_en, C=C_en), axis=1)

# Sort English movies by score in descending order
en_movies = en_movies.sort_values('score', ascending=False)

# Recommend top 10 English movies
en_movies[['title','score']].head(10)
```

# **줄거리 기반 추천 시스템**

```python
# prompt: 'overview' 변수에 대해 TF-IDF 행렬을 만들어줘

# Extract TF-IDF scores for movies with non-empty overviews
df1_df2_merge['overview'] = df1_df2_merge['overview'].fillna('')
tfidf_matrix = tfidf.fit_transform(df1_df2_merge['overview'])
tfidf_matrix.shape
```

## 'The Dark Knight Rises' 영화 줄거리 기반 추천 Top 10

```python
# prompt:  'overview' 변수에 대해 만들어진 TF-IDF 행렬에서 코사인 유사성을 평가해서 유사도 높은 10개 추천 결과를 출력해줘

# Compute cosine similarity matrix
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

# Create a reverse mapping of indices and movie titles
indices = pd.Series(df1_df2_merge.index, index=df1_df2_merge['title']).drop_duplicates()

# Function to get recommendations based on cosine similarity
def get_recommendations(title, cosine_sim=cosine_sim):
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    # Exclude the movie itself from recommendations
    sim_scores = sim_scores[1:11]
    movie_indices = [i[0] for i in sim_scores]
    return df1_df2_merge['title'].iloc[movie_indices]

# Get recommendations for a specific movie (replace 'The Dark Knight Rises' with the movie title you want recommendations for)
recommendations = get_recommendations('The Dark Knight Rises')
print(f"Top 10 Recommendations for 'The Dark Knight Rises':")
recommendations
```

# **메타 데이터 기반 추천**

## **메타 데이터 기반 추천 시스템용 데이터 전처리 과정**

```python
# 각 영화별로 필요한 필드 딕셔너리 생성
def create_metadata_dict(row):
    return {
        'cast': row['cast'],
        'crew': row['crew'],
        'keywords': row['keywords'],
        'genres': row['genres']
    }

# 영화별 딕셔너리 컬럼 추가
df1_df2_merge['metadata'] = df1_df2_merge.apply(create_metadata_dict, axis=1)

def find_similar_movies(target_title, top_n=5):
    # 대상 영화 데이터
    target_movie = df1_df2_merge[df1_df2_merge['title'] == target_title]
    if target_movie.empty:
        return f"영화 '{target_title}'를 찾을 수 없습니다."

    target_metadata = target_movie['metadata'].values[0]

    # 유사도 계산 (단순 비교 예시: 필드 값이 일치하는 영화 찾기)
    def similarity_score(row):
        score = 0
        metadata = row['metadata']
        # 각 필드별 유사도 점수 계산 (단순 문자열 일치 개수 기준)
        for key in ['cast', 'crew', 'director', 'keywords', 'genres']:
            if isinstance(metadata[key], list) and isinstance(target_metadata[key], list):
                intersection = set(metadata[key]) & set(target_metadata[key])
                score += len(intersection)
            elif isinstance(metadata[key], str) and isinstance(target_metadata[key], str):
                if metadata[key] == target_metadata[key]:
                    score += 1
        return score

    # 점수 계산
    df1_df2_merge['similarity'] = df1_df2_merge.apply(similarity_score, axis=1)

    # 대상 영화 제외하고 상위 추천
    recommendations = df1_df2_merge[df1_df2_merge['title'] != target_title] \
        .sort_values(by='similarity', ascending=False).head(top_n)

    return recommendations[['title', 'similarity']]
```

## **Avatar와 유사한 메타 데이터를 가진 영화 추천**

```python
def find_similar_to_avatar(top_n=10):
    # 'Avatar' 영화 데이터 찾기
    avatar_row = df1_df2_merge[df1_df2_merge['title'] == 'Avatar']
    if avatar_row.empty:
        return "영화 'Avatar'를 찾을 수 없습니다."

    avatar_metadata = avatar_row['metadata'].values[0]

    # 유사도 계산 함수
    def similarity_score(row):
        score = 0
        metadata = row['metadata']
        for key in ['cast', 'crew', 'keywords', 'genres']:
            if isinstance(metadata[key], list) and isinstance(avatar_metadata[key], list):
                intersection = set(metadata[key]) & set(avatar_metadata[key])
                score += len(intersection)
            elif isinstance(metadata[key], str) and isinstance(avatar_metadata[key], str):
                if metadata[key] == avatar_metadata[key]:
                    score += 1
        return score

    # 유사도 계산 및 새 컬럼에 저장
    df1_df2_merge['similarity'] = df1_df2_merge.apply(similarity_score, axis=1)

    # 'Avatar' 제외하고 유사도 높은 순으로 정렬 후 상위 10개 선택
    recommendations = df1_df2_merge[df1_df2_merge['title'] != 'Avatar'] \
        .sort_values(by='similarity', ascending=False) \
        .head(top_n)

    # 인덱스 재설정 후 순위 번호 컬럼 추가
    recommendations = recommendations.reset_index(drop=True)
    recommendations['순위'] = recommendations.index + 1

    # '순위'와 'title'만 보여주기
    for idx, row in recommendations.iterrows():
        print(f"{row['순위']}. {row['title']}")

# 호출
print("Avatar와 유사한 영화 추천 목록:")
find_similar_to_avatar(top_n=10)
```
